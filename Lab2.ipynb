```python
def oneof(liter, terminals):
  terms = terminals.split(" ")
  return liter in terms
```

```python
def LL1(word):
  i = 0
  stack = [0]
  accept = 1
  liters = word.split(" ")
  l = 0
  while i != 0 or l < len(liters):
    if accept:
      liter = liters[l]
      l += 1
    if oneof(liter, table["Terminals"][i]):
      accept = table["Accept"][i]
      if table["Return"][i]:
        i = stack[-1]
        stack.pop()
      else:
        next = table['Next'][i] - 1
        if table['Stack'][i]:
          stack.append(i + 1)
        i = next
    elif table['Error'][i]:
      print("Ошибка")
      return
    else:
      i += 1
      accept = 0
    print(liter)
    print(stack)
    print(i)
    print(accept)
    print("---------------------------------------------------------")
  print("Слово принадлежит алфавиту")  
```
